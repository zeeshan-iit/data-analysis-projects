```python
import pandas as pd
```

```python
ds = pd.read_csv("loan.csv")
```

```python
ds.isnull().sum()
```

```python
ds["Gender"] = ds["Gender"].fillna(ds["Gender"].mode()[0])
```

```python
ds["Married"] = ds["Married"].fillna(ds["Married"].mode()[0])
```

```python
req_data = ds[["Gender","Married"]]
```

```python
req_data.isnull().sum()
```

```python
pd.get_dummies(req_data)
```
